```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def plot_metric_oneax(
    ax,
    dfs_s: dict,
    dfs_m: dict,
    metric="rmse",
    title=None,
    show_legend=True,
):
    # 对齐 node 后拼起来
    rows = []
    for name, df_s in dfs_s.items():
        df_m = dfs_m[name]
        a = df_s[["node", metric]].rename(columns={metric: "x"})
        b = df_m[["node", metric]].rename(columns={metric: "y"})
        mm = a.merge(b, on="node", how="inner")
        mm["model"] = name
        rows.append(mm)
    df = pd.concat(rows, ignore_index=True)

    x = df["x"].to_numpy()
    y = df["y"].to_numpy()
    lo = float(np.nanmin(np.r_[x, y]))
    hi = float(np.nanmax(np.r_[x, y]))
    pad = 0.1 * (hi - lo + 1e-12)
    lo, hi = lo - pad, hi + pad

    model_order = list(dfs_s.keys())
    cmap = {m: f"C{i}" for i, m in enumerate(model_order)}
    markers = ["o", "s", "^", "D", "P", "X"]
    mkr = {m: markers[i % len(markers)] for i, m in enumerate(model_order)}

    for m in model_order:
        dd = df[df["model"] == m]
        ax.scatter(
            dd["x"], dd["y"],
            s=50, alpha=0.8,
            color=cmap[m], marker=mkr[m],
            label=m
        )

    ax.plot([lo, hi], [lo, hi], "k--", linewidth=1, label="Reference line (y=x)")
    ax.set_xlim(lo, hi)
    ax.set_ylim(lo, hi)
    ax.grid(True, linestyle=":", alpha=0.5)

    if metric == "mse":
        lab = "MSE"
    elif metric == "rmse":
        lab = "RMSE"
    else:
        lab = "Pinball loss"

    ax.set_xlabel("Metric of Separate forecasting")
    #ax.set_ylabel("Metric of Joint forecasting")
    fig.supylabel("Metric of Joint forecasting", fontsize=24)
    if title is not None:
        ax.set_title(title)

    if show_legend:
        ax.legend()

# ===== 用法：一行三个子图 =====
plt.rcParams.update({
    "axes.titlesize": 24,
    "axes.labelsize": 24,
    "xtick.labelsize": 24,
    "ytick.labelsize": 24,
    "legend.fontsize": 24,
})

dfs_s = {"VAE": dfm_vae_s, "GAN": dfm_gan_s, "Diffusion": dfm_diff_s}
dfs_m = {"VAE": dfm_vae_m, "GAN": dfm_gan_m, "Diffusion": dfm_diff_m}

metrics = ["mse", "rmse", "pinball_avg"]
titles  = ["MSE", "RMSE", "Pinball loss"]

fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharex=False, sharey=False)

for ax, metric, t in zip(axes, metrics, titles):
    plot_metric_oneax(
        ax, dfs_s, dfs_m,
        metric=metric,
        title=t,
        show_legend=False
    )
    ax.set_ylabel("")           # 关键：去掉每个子图的y标签，减少左右留白
    ax.margins(x=0.0, y=0.0)    # 可选：再收一点坐标轴内部留白


# 只放一个总图例（从第一个子图取 handles/labels）
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(
    handles, labels,
    loc="lower center",
    ncol=4,
    frameon=False,
    bbox_to_anchor=(0.5, -0.2)   # 放在图内底部，避免保存时左右/下方留白被撑大
)

# 手动收紧边距与子图间距
fig.subplots_adjust(left=0.07, right=0.995, bottom=0.12, top=0.92, wspace=0.20)

fig.savefig("../Figure/y_x.pdf", bbox_inches="tight", pad_inches=0.02)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def build_improve_box(df_param, df_models: dict, metrics, node_col="node"):
    """
    improve_pct = (parametric - model) / parametric * 100
    >0 表示相对 parametric 有提升（误差下降）
    """
    base = df_param[[node_col] + metrics].copy()

    rows = []
    for model_name, df in df_models.items():
        tmp = df[[node_col] + metrics].merge(base, on=node_col, suffixes=("", "_base"))
        for met in metrics:
            b = tmp[f"{met}_base"].to_numpy(dtype=float)
            m = tmp[met].to_numpy(dtype=float)
            improve = (b - m) / (np.abs(b) ) * 100.0
            rows.append(pd.DataFrame({
                node_col: tmp[node_col],
                "model": model_name,
                "metric": met,
                "improve_pct": improve
            }))
    return pd.concat(rows, ignore_index=True)
def boxplot_improve_1x3(dd, metrics, order=("VAE","GAN","Diffusion"),
                       showfliers=False, save_path=None):
    fig, axes = plt.subplots(1, 3, figsize=(14, 4.6), sharey=False)

    colors = ["C0", "C1", "C2"]
    title_map = {"mse":"MSE", "rmse":"RMSE", "pinball_avg":"Pinball loss"}

    for ax, met in zip(axes, metrics):
        data = [
            dd[(dd["metric"] == met) & (dd["model"] == m)]["improve_pct"].dropna().values
            for m in order
        ]

        bp = ax.boxplot(
            data, labels=order, showfliers=showfliers,
            widths=0.5,  # 箱子窄一点
            patch_artist=True
        )

        for patch, c in zip(bp["boxes"], colors):
            patch.set_facecolor(c); patch.set_alpha(0.25)
            patch.set_edgecolor(c)
        for med, c in zip(bp["medians"], colors):
            med.set_color(c); med.set_linewidth(2)

        ax.axhline(0, color="k", linestyle="--", linewidth=1)
        ax.set_title(title_map.get(met, met))
        ax.grid(True, axis="y", linestyle=":", alpha=0.5)

        # 收紧每个子图左右留白（二选一）
        ax.margins(x=0.02)
        # ax.set_xlim(0.6, len(order) + 0.4)

    axes[0].set_ylabel("Improvement (%)")

    # 收紧整张图左右外边距 + 子图间距
    fig.subplots_adjust(left=0.06, right=0.995, bottom=0.14, top=0.92, wspace=0.25)

    if save_path is not None:
        fig.savefig(save_path, bbox_inches="tight", pad_inches=0.02)

    plt.show()

plt.rcParams.update({
    "axes.titlesize": 24,
    "axes.labelsize": 24,
    "xtick.labelsize": 24,
    "ytick.labelsize": 24,
    "legend.fontsize": 24,
})
df_models = {"VAE": dfm_vae_m, "GAN": dfm_gan_m, "Diffusion": dfm_diff_m}
metrics = ["mse", "rmse", "pinball_avg"]
dd_p = build_improve_box(dfm_parametric, df_models, metrics, node_col="node")
boxplot_improve_1x3(dd_p, metrics, order=("VAE","GAN","Diffusion"), showfliers=False, save_path='../Figure/box_parametric.pdf')

dd_np = build_improve_box(dfm_non_parametric, df_models, metrics, node_col="node")
boxplot_improve_1x3(dd_np, metrics, order=("VAE","GAN","Diffusion"), showfliers=False,save_path='../Figure/box_non_parametric.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from scipy.stats import chi2, pearsonr, gaussian_kde

def plot_joint_ellipsoid_with_marginals(
    packs,
    node_a, node_b,
    time_step=0,
    labels=None,
    node_names=None,
    confidence=0.95,
    figsize=(7.8, 7.2),
    alpha_scatter=0.28,      # 云状更像图1：调低 alpha + 更多点
    s_scatter=10,
    show_truth=True,
    truth_pack_index=0,
    bins=50,                 # 边缘直方图 bins
    hist_alpha=0.35,
    share_marginal_lim=True,
    show_ellipse=True,
    ellipse_lw=2.2,

    # --- KDE smoothing (added) ---
    bw_adjust=1.3,
    kde_lw=2.0,
):
    if not isinstance(packs, (list, tuple)):
        packs = [packs]
    if labels is None:
        labels = [f"Model{i}" for i in range(len(packs))]
    if len(labels) != len(packs):
        raise ValueError("labels 长度必须和 packs 数量一致")

    # --- layout: top hist, right hist, center scatter ---
    fig = plt.figure(figsize=figsize)
    gs = GridSpec(
        2, 2, figure=fig,
        width_ratios=[4.8, 1.4],
        height_ratios=[1.4, 4.8],
        wspace=0.05, hspace=0.05
    )
    ax_top   = fig.add_subplot(gs[0, 0])
    ax_joint = fig.add_subplot(gs[1, 0])
    ax_right = fig.add_subplot(gs[1, 1], sharey=ax_joint)

    # 不显示边缘子图多余刻度
    plt.setp(ax_top.get_xticklabels(), visible=False)
    plt.setp(ax_right.get_yticklabels(), visible=False)
    ax_top.tick_params(axis="x", length=0)
    ax_right.tick_params(axis="y", length=0)

    # truth
    if show_truth:
        Y_true = packs[truth_pack_index]["Y_true"]
        if hasattr(Y_true, "cpu"): Y_true = Y_true.cpu().numpy()
        if Y_true.ndim == 3: Y_true = Y_true[0]

        true_a = float(Y_true[node_a, time_step])
        true_b = float(Y_true[node_b, time_step])
        ax_joint.scatter(true_a, true_b, c="black", s=220, marker="*",
                         edgecolors="white", linewidths=1.2, zorder=10, label="Ground Truth")

    # ellipse common
    scale = np.sqrt(chi2.ppf(confidence, df=2))
    t = np.linspace(0, 2*np.pi, 240)
    circle = np.vstack([np.cos(t), np.sin(t)])

    # 为了让边缘直方图范围一致，可先收集所有样本
    all_a, all_b = [], []
    for pack in packs:
        Y_pred = pack["Y_pred"]
        if hasattr(Y_pred, "cpu"): Y_pred = Y_pred.cpu().numpy()
        all_a.append(Y_pred[:, node_a, time_step])
        all_b.append(Y_pred[:, node_b, time_step])
    all_a = np.concatenate(all_a)
    all_b = np.concatenate(all_b)

    if share_marginal_lim:
        ax_joint.set_xlim(all_a.min(), all_a.max())
        ax_joint.set_ylim(all_b.min(), all_b.max())
        ax_top.set_xlim(ax_joint.get_xlim())
        ax_right.set_ylim(ax_joint.get_ylim())

    # plot each pack
    for i, pack in enumerate(packs):
        lab = labels[i]
        Y_pred = pack["Y_pred"]
        if hasattr(Y_pred, "cpu"): Y_pred = Y_pred.cpu().numpy()

        samp_a = Y_pred[:, node_a, time_step]
        samp_b = Y_pred[:, node_b, time_step]
        corr_val, _ = pearsonr(samp_a, samp_b)

        # joint scatter
        sc = ax_joint.scatter(
            samp_a, samp_b,
            s=s_scatter, alpha=alpha_scatter,
            edgecolors="none",
            label=f"samples (corr={corr_val:.2f})"
        )

        # marginals (同色)
        color = sc.get_facecolors()[0]
        ax_top.hist(samp_a, bins=bins, density=True, alpha=hist_alpha, color=color)
        ax_right.hist(samp_b, bins=bins, density=True, alpha=hist_alpha, color=color,
                      orientation="horizontal")

        # --- KDE overlay (smooth) ---
        # top (x-marginal)
        xmin, xmax = ax_top.get_xlim()

        grid_x = np.linspace(xmin, xmax, 300)
        kde_x = gaussian_kde(samp_a)
        kde_x.set_bandwidth(kde_x.factor * bw_adjust)
        dens_x = kde_x(grid_x)
        ax_top.plot(grid_x, dens_x, color=color, lw=kde_lw)

        # right (y-marginal)
        ymin, ymax = ax_right.get_ylim()
        grid_y = np.linspace(ymin, ymax, 300)
        kde_y = gaussian_kde(samp_b)
        kde_y.set_bandwidth(kde_y.factor * bw_adjust)
        dens_y = kde_y(grid_y)
        ax_right.plot(dens_y, grid_y, color=color, lw=kde_lw)

        # ellipse
        if show_ellipse:
            mu = np.array([np.mean(samp_a), np.mean(samp_b)])
            cov = np.cov(samp_a, samp_b)
            eig_vals, eig_vecs = np.linalg.eigh(cov)
            eig_vals = np.maximum(eig_vals, 1e-12)
            A = eig_vecs @ np.diag(np.sqrt(eig_vals) * scale)
            ell = mu.reshape(2, 1) + A @ circle
            ax_joint.plot(ell[0], ell[1], lw=ellipse_lw, color=color,
                          label=f"{int(confidence*100)}% quantiles")

    name_a = f"Node {node_a}" if node_names is None else node_names[node_a]
    name_b = f"Node {node_b}" if node_names is None else node_names[node_b]
    ax_joint.set_xlabel(f"Load of {name_a}")
    ax_joint.set_ylabel(f"Load of {name_b}")
    ax_joint.grid(True, alpha=0.25, linestyle="--")
    #ax_top.set_title(f"Joint @ time_step={time_step}")

    # marginal cosmetics
    ax_top.grid(True, alpha=0.12, linestyle="--")
    ax_right.grid(True, alpha=0.12, linestyle="--")

    # legend 放 joint 里
    ax_joint.legend(loc="best", frameon=True)

    plt.tight_layout()
    return fig, (ax_top, ax_joint, ax_right)
```

```python
import matplotlib.pyplot as plt

eps_cols = ['10.0', '100.0', '1000.0', '10000.0']
df=pd.read_csv('../cost on validation with different eps.csv')
df['model']=['Diffusion_Joint','Diffusion_Separate','GAN_Joint','GAN_Separate','Non_parametric','Parametric','VAE_Joint','VAE_Separate']
df_long = df.melt(
    id_vars="model",
    value_vars=eps_cols,
    var_name="eps",
    value_name="cost"
)
df_long["eps"] = df_long["eps"].astype(float)

# 分组：只有 diffusion / gan / vae 是成组的
def get_group(model):
    if model.startswith("Diffusion"):
        return "Diffusion"
    elif model.startswith("GAN"):
        return "GAN"
    elif model.startswith("VAE"):
        return "VAE"
    elif model == "Parametric":
        return "Parametric"
    elif model == "Non_parametric":
        return "Non_parametric"
    else:
        return "Other"

# m / s 区分；单独模型给自己一种样式
def get_variant(model):
    if model.endswith("_Joint"):
        return "Joint"
    elif model.endswith("_Separate"):
        return "Separate"
    elif model == "Parametric":
        return "Single_p"
    elif model == "Non_parametric":
        return "Single_np"
    else:
        return "Other"

df_long["group"] = df_long["model"].apply(get_group)
df_long["variant"] = df_long["model"].apply(get_variant)

# 同组同色
color_map = {
    "Diffusion": "#4C78A8",
    "GAN": "#54A24B",
    "VAE": "#F58518",
    "Parametric": "#B279A2",
    "Non_parametric": "#E45756",
    "Other": "#9D9D9D"
}

# 样式区分
marker_map = {
    "Joint": "o",
    "Separate": "s",
    "Single_p": "D",
    "Single_np": "D",
    "Other": "o"
}

linestyle_map = {
    "Joint": "-",
    "Separate": "--",
    "Single_p": "-.",
    "Single_np": "-.",
    "Other": "-"
}

# 画布更大
fig, ax = plt.subplots(figsize=(12.5, 7.5), dpi=200)

for model, sub in df_long.groupby("model"):
    group = sub["group"].iloc[0]
    variant = sub["variant"].iloc[0]

    ax.plot(
        sub["eps"],
        sub["cost"],
        label=model,
        color=color_map[group],
        linestyle=linestyle_map[variant],
        marker=marker_map[variant],
        linewidth=2.8,
        markersize=8.5,
        alpha=0.95
    )

ax.set_xscale("log")

# 字体调大
ax.set_xlabel("Radius of Ambiguity Set", fontsize=20, labelpad=10)
ax.set_ylabel("Average Cost (USD)", fontsize=20, labelpad=10)
ax.set_title("Average Cost vs Radius of Ambiguity Set", fontsize=20, pad=20)

# tick 字体放大
ax.tick_params(axis="both", which="major", labelsize=18)

# 美化
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.grid(True, which="major", linestyle="--", alpha=0.25)

# 图例更大
ax.legend(
    title=None,
    bbox_to_anchor=(0.5, -0.2),
    loc="upper center",
    ncol=4,
    frameon=False,
    fontsize=18,
    handlelength=2.4,
    columnspacing=1.8
)

plt.tight_layout()
plt.savefig('../Figure/val_cost_with_eps.pdf')
plt.show()
```
